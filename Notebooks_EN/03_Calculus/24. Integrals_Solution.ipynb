$$\int x e^x \, dx = x e^x - \int e^x dx = x e^x - e^x + C$$

The integral of $f(x) = 2x + 1$ is:  
$$
\int (2x + 1) \, dx = x^2 + x + C
$$

Evaluate over $[0, \pi]$:  
$$
\int_{0}^{\pi} (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi
$$

Substitute the limits:  
$$
\int_{0}^{\pi} (2x + 1) \, dx = (\pi^2 + \pi) - (0^2 + 0)
$$
$$
\int_{0}^{\pi} (2x + 1) \, dx = \pi^2 + \pi
$$


---

#### 2. Integral of $g(x) = x^2$

The integral of $g(x) = x^2$ is:  
$$
\int x^2 \, dx = \frac{x^3}{3} + C
$$

Evaluate over $[0, \pi]$:  
$$
\int_{0}^{\pi} x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^\pi
$$

Substitute the limits:  
$$
\int_{0}^{\pi} x^2 \, dx = \frac{\pi^3}{3} - \frac{0^3}{3}
$$
$$
\int_{0}^{\pi} x^2 \, dx = \frac{\pi^3}{3}
$$

---

### Final Results
1. The integral of $f(x) = 2x + 1$ over $[0, \pi]$ is:

\int_{0}^{\pi} (2x + 1) \, dx = \pi^2 + \pi

### Solution: Area of the Region Bounded by $x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$

#### 1. Define the Bounded Region
The region is enclosed by:
- Vertical lines: $x = 1$ and $x = 2$
- Horizontal line: $y = 0$ (x-axis)
- Curve: $y = x^2 + 1$

$$
\text{Area} = \int_{1}^{2} (x^2 + 1) dx
$$

---

#### 2. Compute the Integral

**Step 1. Setup the integral:**
$$
\text{Area} = \int_{1}^{2} (x^2 + 1) dx
$$

**Step 2. Find the antiderivative:**
$$
\int (x^2 + 1) dx = \frac{x^3}{3} + x + C
$$

**Step 3. Evaluate over $[1, 2]$:**
$$
\text{Area} = \left[ \frac{x^3}{3} + x \right]_1^2
$$

**Step 4. Substitute the limits:**
$$
\text{Area} = \left( \frac{2^3}{3} + 2 \right) - \left( \frac{1^3}{3} + 1 \right)
$$
$$
\text{Area} = \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right)
$$
$$
\text{Area} = \frac{8}{3} + \frac{6}{3} - \frac{1}{3} - \frac{3}{3}
$$
$$
\text{Area} = \frac{10}{3}
$$

---

### Final Result
The area of the region is:
$$
\text{Area} = \frac{10}{3} \approx 3.33
$$

# Exercise 6

#### Step 1: Compute the Velocity Function
Given:
$$
x(t) = 3t^2 - 6t + 1
$$
Differentiate:
$$
v(t) = \frac{dx}{dt} = 6t - 6
$$



#### Step 2: Solve for Critical Points of $v(t)$
Solve $v(t) = 0$:
$$
6t - 6 = 0 \quad \implies \quad t = 1
$$



#### Step 3: Split the Integral
The distance is:
$$
\text{Distance} = \int_0^1 |v(t)| \, dt + \int_1^2 |v(t)| \, dt
$$
On $[0, 1]$, $v(t) < 0$, so $|v(t)| = -v(t) = -6t + 6$.  
On $[1, 2]$, $v(t) > 0$, so $|v(t)| = v(t) = 6t - 6$.



#### Step 4: Evaluate Each Integral

**First Integral ($[0, 1]$):**
$$
\int_0^1 (-6t + 6) \, dt = \left[ -3t^2 + 6t \right]_0^1
$$
$$
= \left(-3(1)^2 + 6(1)\right) - \left(-3(0)^2 + 6(0)\right)
$$
$$
= -3 + 6 = 3
$$

**Second Integral ($[1, 2]$):**
$$
\int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2
$$
$$
= \left(3(2)^2 - 6(2)\right) - \left(3(1)^2 - 6(1)\right)
$$
$$
= \left(12 - 12\right) - \left(3 - 6\right)
$$
$$
= 0 - (-3) = 3
$$



#### Step 5: Total Distance
Add the results:
$$
\text{Distance} = 3 + 3 = 6
$$


### Final Result
The total distance traveled by the particle is:
$$
\text{Distance} = 6
$$
