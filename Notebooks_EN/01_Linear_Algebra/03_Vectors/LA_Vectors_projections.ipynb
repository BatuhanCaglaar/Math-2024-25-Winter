# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

### Problem: Divide the Line Segment in a Given Ratio

To solve this problem, I will divide the line segment joining points $A(-1, 2)$ and $B(3, -2)$ in the ratio $1:3$. Then, I will illustrate the result on a graph.


### Step 1: Understand the Section Formula

The **section formula** is used to find the coordinates of a point $P(x, y)$ that divides a line segment joining two points $A(x_1, y_1)$ and $B(x_2, y_2)$ in a given ratio $m:n$.

The formula is:
$$
P(x, y) = \left( \frac{mx_2 + nx_1}{m + n}, \frac{my_2 + ny_1}{m + n} \right)
$$

In this problem:
- $A(x_1, y_1) = (-1, 2)$
- $B(x_2, y_2) = (3, -2)$
- The ratio is $m:n = 1:3$.


### Step 2: Apply the Section Formula

#### Find the $x$-coordinate of $P$:
$$
x = \frac{m \cdot x_2 + n \cdot x_1}{m + n}
$$

Substituting $m = 1$, $n = 3$, $x_1 = -1$, $x_2 = 3$:
$$
x = \frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3} = \frac{3 - 3}{4} = 0
$$

#### Find the $y$-coordinate of $P$:
$$
y = \frac{m \cdot y_2 + n \cdot y_1}{m + n}
$$

Substituting $m = 1$, $n = 3$, $y_1 = 2$, $y_2 = -2$:
$$
y = \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3} = \frac{-2 + 6}{4} = \frac{4}{4} = 1
$$

Thus, the coordinates of the dividing point $P$ are:
$$
P(0, 1)
$$



### Step 3: Verify the Solution

The point $P(0, 1)$ divides the line segment $AB$ in the ratio $1:3$. The calculations confirm the accuracy.


### Step 4: Illustrate on a Graph

Next, I will plot the points $A(-1, 2)$, $B(3, -2)$, and $P(0, 1)$ on a graph. This will show that $P$ divides $AB$ in the given ratio.

# Problem: Projection of Vector $\mathbf{a} = (3, 4)$ onto the $OX$ and $OY$ Axes

## Step 1: Understanding Vector Projection

To project a vector onto an axis, we use the following principles:

1. **Projection onto $OX$ (horizontal axis):**
   - Retain the $x$-component, set the $y$-component to $0$.
   - Formula: $\text{Projection on } OX = (x, 0)$.

2. **Projection onto $OY$ (vertical axis):**
   - Retain the $y$-component, set the $x$-component to $0$.
   - Formula: $\text{Projection on } OY = (0, y)$.

Given $\mathbf{a} = (3, 4)$:
- $x$-component: $3$
- $y$-component: $4$

## Step 2: Calculate the Projections

### Projection onto $OX$:
Using the formula:
$$
\text{Projection on } OX = (3, 0)
$$

### Projection onto $OY$:
Using the formula:
$$
\text{Projection on } OY = (0, 4)
$$

Thus:
- The vector projected onto the $OX$ axis is $(3, 0)$.
- The vector projected onto the $OY$ axis is $(0, 4)$.

## Step 3: Explanation

- The projection onto $OX$ shows that the shadow of the vector $\mathbf{a}$ on the horizontal axis lies at $(3, 0)$.
- The projection onto $OY$ shows that the shadow of the vector $\mathbf{a}$ on the vertical axis lies at $(0, 4)$.


```python
import matplotlib.pyplot as plt #Exercise 3
import numpy as np

# Vektörler
a = np.array([2, 3])
b = np.array([1, 1])

# Projeksiyonu hesapla
dot_product = np.dot(a, b)
norm_b_squared = np.dot(b, b)
projection = (dot_product / norm_b_squared) * b

# Grafik oluştur
plt.figure(figsize=(6, 6))

# Orijinal vektörler
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (2, 3)')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b (1, 1)')

# Projeksiyon
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='red', label='Projection of a onto b')

# Eksenler ve ayarlar
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.title('Projection of Vector (2, 3) onto Vector (1, 1)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Problem: Project Vector $\mathbf{a} = (2, 3)$ onto Vector $\mathbf{b} = (1, 1)$

Let me walk you through the process of projecting vector $\mathbf{a}$ onto $\mathbf{b}$. We use the formula for vector projection:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{b}\|^2} \mathbf{b}
$$

This formula allows us to find the projection of one vector onto another.



## Step 1: Identify the Components
We are given:
- $\mathbf{a} = (2, 3)$
- $\mathbf{b} = (1, 1)$



## Step 2: Calculate the Dot Product
The dot product $\mathbf{a} \cdot \mathbf{b}$ is defined as:

$$
\mathbf{a} \cdot \mathbf{b} = a_x b_x + a_y b_y
$$

Substituting the components of $\mathbf{a}$ and $\mathbf{b}$:

$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 2 + 3 = 5
$$



## Step 3: Find the Magnitude of $\mathbf{b}$
The magnitude of $\mathbf{b}$ is calculated as:

$$
\|\mathbf{b}\| = \sqrt{b_x^2 + b_y^2}
$$

Substituting the components of $\mathbf{b}$:

$$
\|\mathbf{b}\| = \sqrt{1^2 + 1^2} = \sqrt{1 + 1} = \sqrt{2}
$$

The square of the magnitude is:

$$
\|\mathbf{b}\|^2 = (\sqrt{2})^2 = 2
$$



## Step 4: Apply the Projection Formula
Now we substitute the values into the projection formula:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{b}\|^2} \mathbf{b}
$$

Substitute $\mathbf{a} \cdot \mathbf{b} = 5$ and $\|\mathbf{b}\|^2 = 2$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \mathbf{b}
$$

Since $\mathbf{b} = (1, 1)$, we calculate:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} (1, 1)
$$

Performing the scalar multiplication:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \left(\frac{5}{2}, \frac{5}{2}\right)
$$



## Final Result
The projection of $\mathbf{a}$ onto $\mathbf{b}$ is:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \left(\frac{5}{2}, \frac{5}{2}\right)
$$
