# Exercises

* Find the equation of a circle with center at point $A(1,2)$ and radius $r=3$.
* Find the equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through point $y(3)=1$.
* Find the center of the ellipse with the equation $x^2 + 4y^2 - 4x - 16y + 16 = 0$.
* Find the slope ($m>0$) of the line $y=mx-5$ that is tangent to the circle with the equation $x^2 + y^2=1$.
* Find the intersection points of the hyperbola $x^2 - y^2 = 1$ with the ellipse's line $x^2 + 4y^2 = 6$.
* For the given hyperbola $x^2 - y^2 = 1$, find the distance between its branches.

```python
import numpy as np  #Exercise 1
import matplotlib.pyplot as plt

# Circle center and radius
center = (1, 2)  # Center (x₀, y₀)
radius = 3       # Radius (r)

# Circle equation: (x - x₀)² + (y - y₀)² = r²
# This can be expressed parametrically:
# x = x₀ + r*cos(t), y = y₀ + r*sin(t)

# Parametric angle range
theta = np.linspace(0, 2 * np.pi, 500)

# Calculate points on the circle
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

# Plotting the circle
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=f'Circle: (x - {center[0]})² + (y - {center[1]})² = {radius**2}')
plt.scatter(center[0], center[1], color='red', label=f'Center: {center}')
plt.gca().set_aspect('equal', adjustable='box')  # Ensures the circle looks proper
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Circle Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Step-by-Step Solution

#### General Equation of a Circle:
The equation of a circle is:

$$(x - h)^2 + (y - k)^2 = r^2$$

#### Substituting the values:
- Center: $A(1, 2)$, so $h = 1$ and $k = 2$
- Radius: $r = 3$, so $r^2 = 9$

The equation becomes:

$$(x - 1)^2 + (y - 2)^2 = 9$$

#### Expanded Form (Optional):
Expanding $(x - 1)^2$ and $(y - 2)^2$:

$$(x - 1)^2 = x^2 - 2x + 1$$  
$$(y - 2)^2 = y^2 - 4y + 4$$  

Substitute these into the equation:

$$x^2 - 2x + 1 + y^2 - 4y + 4 = 9$$

Combine like terms:

$$x^2 + y^2 - 2x - 4y + 5 = 9$$

Simplify further:

$$x^2 + y^2 - 2x - 4y - 4 = 0$$

#### Final Answer:
- Standard Form:  
  $$(x - 1)^2 + (y - 2)^2 = 9$$
- Expanded Form:  
  $$x^2 + y^2 - 2x - 4y - 4 = 0$$

```python
import numpy as np #Exercise 2
import matplotlib.pyplot as plt

# Step 1: Define the points
x1, y1 = 2, 0  # Point A (root)
x2, y2 = 4, 0  # Point B (root)
x3, y3 = 3, 1  # Point C (parabola passes through)

# Step 2: Calculate the coefficient 'a' using point C
# General form of the parabola: f(x) = a * (x - x1) * (x - x2)
a = y3 / ((x3 - x1) * (x3 - x2))

# Step 3: Define the function for the parabola
def parabola(x):
    return a * (x - x1) * (x - x2)

# Step 4: Generate x values and compute y values
x_vals = np.linspace(1, 5, 500)  # x values from 1 to 5
y_vals = parabola(x_vals)

# Step 5: Plot the parabola and points
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=f"Parabola: y = {a:.2f}(x - {x1})(x - {x2})", color='blue')
plt.scatter([x1, x2, x3], [y1, y2, y3], color='red', label='Given Points')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # y-axis

# Annotate points
plt.text(x1, y1 - 0.3, f"A({x1}, {y1})", fontsize=10, ha='center')
plt.text(x2, y2 - 0.3, f"B({x2}, {y2})", fontsize=10, ha='center')
plt.text(x3, y3 + 0.3, f"C({x3}, {y3})", fontsize=10, ha='center')

# Add labels and legend
plt.title("Parabola Passing Through Points A(2,0), B(4,0), and C(3,1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

Since the parabola intersects the $Ox$ axis at $x = 2$ and $x = 4$, its general equation can be written as:

$$
y = a(x - 2)(x - 4)
$$


### Step 2: Use the Given Point to Determine $a$

$$
1 = a(3 - 2)(3 - 4)
$$


$$
1 = a(1)(-1)
$$

$$
1 = -a
$$


$$
a = -1
$$



### Step 3: Write the Final Equation


$$
y = -1(x - 2)(x - 4)
$$


$$
y = -(x^2 - 6x + 8)
$$


$$
y = -x^2 + 6x - 8
$$



### Final Answer


$$
y = -x^2 + 6x - 8

```python
import sympy as sp #Exercise 3

# Define symbols
x, y = sp.symbols('x y')
```

```python
# Define the equation of the ellipse
equation = x**2 + 4*y**2 - 4*x - 16*y + 16
```

```python
# Rearrange the equation for completing the square
eq_x = x**2 - 4*x  # x terms
eq_y = 4*y**2 - 16*y  # y terms
```

```python
# Complete the square for x and y terms
completed_x = sp.simplify(eq_x + 4)  # Adding and subtracting 4 to complete the square
completed_y = sp.simplify(eq_y - 16)  # Adding and subtracting 16 to complete the square
```

```python
# Simplify the full equation
simplified_eq = completed_x + completed_y + 16 - 16  # Restoring the constants
simplified_eq = sp.simplify(simplified_eq)
```

```python
# The center of the ellipse is given by the negative of the linear terms in the completed squares
center_x = 4 / 2  # -(-4)/2
center_y = 16 / 8  # -(-16)/8
# Display results
print(f"Center of the ellipse: ({center_x}, {center_y})")
```

```python
import sympy as sp #Exercise 4

x, m = sp.symbols('x m')  # 'x' for the x-coordinate, 'm' for the slope of the line
```

```python
y = m * x - 5  # The equation of the line in terms of x and m
```

```python
equation = x**2 + y**2 - 1  # The equation of the circle
```

```python
equation = equation.subs(y, m*x - 5)  # Substitute y = mx - 5 into the equation of the circle
```

```python
simplified_equation = sp.simplify(equation)
```

```python
quadratic_equation = sp.expand(simplified_equation)
```

```python
coeffs = sp.Poly(quadratic_equation, x).coeffs()
a = coeffs[0]  # Coefficient of x^2
b = coeffs[1]  # Coefficient of x
c = coeffs[2]  # Constant term
```

```python
# Set the discriminant to 0 for tangency
discriminant = b**2 - 4*a*c
discriminant_condition = sp.Eq(discriminant, 0)
```

```python
# Solve for the slope 'm'
m_values = sp.solve(discriminant_condition, m)
m_values
```

```python
import sympy as sp
x, y = sp.symbols('x y')
```

```python
# Define the equations for the hyperbola and ellipse
hyperbola_eq = x**2 - y**2 - 1  # x^2 - y^2 = 1
ellipse_eq = x**2 + 4*y**2 - 6  # x^2 + 4y^2 = 6
```

```python
# Solve the system of equations
intersection_points = sp.solve([hyperbola_eq, ellipse_eq], (x, y))
intersection_points
```

### Solution

1. Hyperbola: $ x^2 - y^2 = 1 $
2. Ellipse: $ x^2 + 4y^2 = 6 $

#### Step 1: Solve for $x^2$ in terms of $y^2$ from the hyperbola equation:

$$
x^2 = y^2 + 1
$$


$$
(y^2 + 1) + 4y^2 = 6
$$

$$
5y^2 + 1 = 6 \quad \Rightarrow \quad 5y^2 = 5 \quad \Rightarrow \quad y^2 = 1
$$

#### Step 2: Find $x$

- When $y = 1$:
  $$
  x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
  $$
- When $y = -1$:
  $$
  x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
  $$

#### Step 3: Intersection Points

$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$

### Conclusion
The intersection points of the hyperbola $x^2 - y^2 = 1$ and the ellipse $x^2 + 4y^2 = 6$ are:
$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$


```python
import numpy as np  #Exercise 6
import matplotlib.pyplot as plt

# Function for the hyperbola
def hyperbola(x):
    return np.sqrt(x**2 - 1)

# Generate x values for plotting
x_vals = np.linspace(-3, 3, 400)

# Calculate y values for the upper and lower branches of the hyperbola
y_vals_upper = hyperbola(x_vals)
y_vals_lower = -hyperbola(x_vals)

# Plot the hyperbola
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals_upper, label=r'$\mathbf{x^2 - y^2 = 1}$ (Upper branch)', color='blue')
plt.plot(x_vals, y_vals_lower, label=r'$\mathbf{x^2 - y^2 = 1}$ (Lower branch)', color='blue')

# Plot the asymptotes y = x and y = -x
plt.plot(x_vals, x_vals, label=r'$\mathbf{y = x}$ (Asymptote)', color='red', linestyle='--')
plt.plot(x_vals, -x_vals, label=r'$\mathbf{y = -x}$ (Asymptote)', color='red', linestyle='--')

# Mark the intersection points at y = 0
plt.scatter([1, -1], [0, 0], color='green', zorder=5)
plt.text(1, 0.2, '(1, 0)', fontsize=12, horizontalalignment='center')
plt.text(-1, 0.2, '(-1, 0)', fontsize=12, horizontalalignment='center')

# Set plot limits
plt.xlim([-3, 3])
plt.ylim([-3, 3])

# Adding labels and title
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(r'Hyperbola and its Asymptotes $\mathbf{x^2 - y^2 = 1}$')
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Calculate the distance between the branches at y = 0
# The branches intersect the x-axis at x = 1 and x = -1
distance = 1 - (-1)  # Distance between (1, 0) and (-1, 0)
print(f"The distance between the branches of the hyperbola at y = 0 is: {distance}")
```
